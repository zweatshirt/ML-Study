```python
# my notes about covariance and correlation because I haven't taken stats yet lol
# in the process I wanted to roughly cover mean, standard deviation, etc so it gets mathy
%matplotlib inline
import pandas as pd
import seaborn as sb
from math import sqrt

sb.set()
# my attemt to find a correlation on fare cost and age
df = pd.read_csv('titanic.csv')
df.sort_values(["PassengerId"])
ages_and_fares = df[['Fare', 'Age']].dropna()
ages = ages_and_fares['Fare'].to_list()
fares = ages_and_fares['Age'].to_list()
```

```python
# I could be edgy and try to find some correlation between survival and class,
# but that is likely better off with a chi-square test.
# Instead I'll do fare cost and age

def mean(data):
    total = 0
    for i in data: # could've just used sum but wanted to show
        total += i
    return total/len(data)

# 'how spread out data is'
def variance(data):
    k = mean(data) 
    squared_diffs = [(i - k)**2 for i in data]
    n = len(data)
    return sum(squared_diffs) / (n - 1)

def standard_dev(data):
    return sqrt(variance(data))

def dot_prod(a, b):
    # need to 'zip' vals, there's the built in but wanted to do it
    if len(a) < len(b):
        for i in range(len(a) - 1):
            a[i] = a[i] * b[i]
        return sum(a)
    else:
        for i in range(len(b) - 1):
            b[i] = b[i] * a[i]
        return sum(b)
    
def covariance(a, b):
    k_a, k_b = mean(a), mean(b)
    d = dot_prod([i - k_a for i in a], [i - k_b for i in b])
    n = len(a)
    return d / (n - 1)

def correlation(a, b):
    return covariance(a, b) / (standard_dev(a) * standard_dev(b))
```

```python
# scatter plot the data
sb.scatterplot(x='Age', y='Fare', data=df)
```

```python
# But do they have a correlation?
correlation(ages, fares)
# our data says barely any correlation at all, the scatter plot shows this
```

```python
# I think it's safe to say the Seaborn linear regression agrees with this
sb.lmplot(x="Age", y="Fare", data=df)
```
